```python
import numpy as np

# ---------------------------------------------------------
# Helper Functions
# ---------------------------------------------------------

def precision_at_k(actual, predicted, k):
    """
    actual: set/list of true relevant items
    predicted: list of recommended items (ranked)
    """
    predicted_k = predicted[:k]
    relevant = set(actual)
    hit_count = sum([1 for item in predicted_k if item in relevant])
    return hit_count / k


def recall_at_k(actual, predicted, k):
    predicted_k = predicted[:k]
    relevant = set(actual)
    hit_count = sum([1 for item in predicted_k if item in relevant])
    return hit_count / len(relevant) if len(relevant) > 0 else 0


def dcg_at_k(predicted, actual, k):
    predicted_k = predicted[:k]
    dcg = 0.0
    for i, item in enumerate(predicted_k):
        if item in actual:
            dcg += 1 / np.log2(i + 2)      # position i → log2(i+2)
    return dcg


def ndcg_at_k(actual, predicted, k):
    dcg = dcg_at_k(predicted, actual, k)
    ideal_dcg = dcg_at_k(actual, actual, k)   # best-case ranking
    return dcg / ideal_dcg if ideal_dcg > 0 else 0


# ---------------------------------------------------------
# Example Evaluation on Your Recommendations
# ---------------------------------------------------------

# Suppose for a user:
actual_items = [10, 20, 30, 40]        # items user actually interacted with
predicted_items = [20, 50, 10, 60, 70, 30]  # items recommended by your model

for k in [5, 10, 20]:
    print(f"Precision@{k}:", precision_at_k(actual_items, predicted_items, k))
    print(f"Recall@{k}:", recall_at_k(actual_items, predicted_items, k))
    print(f"NDCG@{k}:", ndcg_at_k(actual_items, predicted_items, k))
    print("----")
```

```python
import pandas as pd
import numpy as np

# Load movies dataset
movies = pd.read_csv(r"C:\Users\Hp\OneDrive\Desktop\recommender system\data\tmdb_5000_movies.csv")

# Check columns
print(movies.columns)

# -----------------------------
# 1. Build Popularity Map
# -----------------------------
# Using 'vote_count' as popularity
popularity_map = dict(zip(movies['id'], movies['vote_count']))

# -----------------------------
# 2. Example Recommendations
# -----------------------------
# Suppose these are top-K recommended movie IDs for users
user_recommendations = {
    1: [1, 2, 3, 4, 5],
    2: [10, 20, 30, 40, 50],
    3: [100, 200, 300, 400, 500]
}

# -----------------------------
# 3. ARP & Popularity Lift
# -----------------------------
def avg_rec_popularity(recommended_items, popularity_map):
    pops = [popularity_map.get(item, 0) for item in recommended_items]
    return np.mean(pops)

def popularity_lift(recommended_items, popularity_map):
    arp = avg_rec_popularity(recommended_items, popularity_map)
    catalog_mean = np.mean(list(popularity_map.values()))
    lift = arp / catalog_mean if catalog_mean > 0 else 0
    return arp, catalog_mean, lift

# Compute for all users
user_arp = {}
for user, items in user_recommendations.items():
    arp, catalog_mean, lift = popularity_lift(items, popularity_map)
    user_arp[user] = {'ARP': arp, 'Lift': lift}

# -----------------------------
# 4. Order Users by Diversity
# -----------------------------
sorted_users = sorted(user_arp.items(), key=lambda x: x[1]['ARP'])

print("User ID → ARP, Popularity Lift (lower ARP = more diverse)")
for user, metrics in sorted_users:
    print(user, "→ ARP:", round(metrics['ARP'],2), "Lift:", round(metrics['Lift'],2))
```

```python
import pandas as pd
import numpy as np

# -------------------------------
# Step 1: Load dataset
# -------------------------------
movies = pd.read_csv('tmdb_5000_movies.csv')

# -------------------------------
# Step 2: Bayesian rating
# -------------------------------
C = movies['vote_average'].mean()
m = 50

def bayesian_rating(x, m=m, C=C):
    v = x['vote_count']
    R = x['vote_average']
    return (v/(v+m) * R) + (m/(v+m) * C)

movies['bayesian_score'] = movies.apply(bayesian_rating, axis=1)

# -------------------------------
# Step 3: Top movies by bayesian_score
# -------------------------------
movies_sorted = movies.sort_values('bayesian_score', ascending=False)
top_movies = movies_sorted['id'].tolist()  # full list sorted

# -------------------------------
# Step 4: Evaluation metrics
# -------------------------------
def precision_at_k(actual, predicted, k):
    predicted_k = predicted[:k]
    relevant = set(actual)
    hits = sum([1 for p in predicted_k if p in relevant])
    return hits / k

def recall_at_k(actual, predicted, k):
    predicted_k = predicted[:k]
    relevant = set(actual)
    hits = sum([1 for p in predicted_k if p in relevant])
    return hits / len(relevant) if relevant else 0

def ndcg_at_k(actual, predicted, k):
    predicted_k = predicted[:k]
    dcg = 0
    for i, p in enumerate(predicted_k):
        if p in actual:
            dcg += 1 / np.log2(i + 2)
    ideal_dcg = sum([1/np.log2(i+2) for i in range(min(len(actual), k))])
    return dcg / ideal_dcg if ideal_dcg > 0 else 0

# -------------------------------
# Step 5: User test items (example)
# -------------------------------
user_test_items = [19995, 206647, 1452, 285, 49026]  # your specified movies
print("User liked movies:", user_test_items)

# -------------------------------
# Step 6: Compute for K=5,10,20
# -------------------------------
for K in [5, 10, 20]:
    precision = precision_at_k(user_test_items, top_movies, K)
    recall = recall_at_k(user_test_items, top_movies, K)
    ndcg = ndcg_at_k(user_test_items, top_movies, K)
    print(f"--- K={K} ---")
    print(f"Precision@{K}: {precision:.3f}")
    print(f"Recall@{K}: {recall:.3f}")
    print(f"NDCG@{K}: {ndcg:.3f}\n")
```

```python
import pandas as pd
import numpy as np

# ---------------------------------------------------
# LOAD DATA
# ---------------------------------------------------
movies = pd.read_csv("tmdb_5000_movies.csv")

# ---------------------------------------------------
# GLOBAL VALUES FOR WEIGHTED RATING FORMULA
# ---------------------------------------------------
C = movies['vote_average'].mean()        # global average rating
m = movies['vote_count'].quantile(0.70)  # vote threshold

# ---------------------------------------------------
# WEIGHTED RATING FORMULA (TID FORMULA)
# Formula = (v*R/(v+m)) + (m*C/(v+m))
# ---------------------------------------------------
def weighted_rating(x, m=m, C=C):
    v = x['vote_count']
    R = x['vote_average']
    return (v * R / (v + m)) + (m * C / (v + m))

# Add weighted rating column
movies['weighted_rating'] = movies.apply(weighted_rating, axis=1)

# ---------------------------------------------------
# GINI INDEX FUNCTION
# ---------------------------------------------------
def gini_from_array(x):
    x = np.asarray(x, dtype=np.float64).flatten()

    if x.size == 0:
        return np.nan
    
    # If values are negative → shift
    if x.min() < 0:
        x = x - x.min()
    
    # If all values zero
    if np.allclose(x, 0):
        return 0.0
    
    # Sort values
    x = np.sort(x)
    n = x.size
    cumw = np.arange(1, n + 1) * x
    sum_x = x.sum()
    
    gini = (2.0 * cumw.sum()) / (n * sum_x) - (n + 1.0) / n
    return float(np.clip(gini, 0.0, 1.0))

def gini_summary(name, series):
    g = gini_from_array(series)
    print(f"{name}: Gini = {g:.6f}, Mean = {np.mean(series):.6f}")
    return g

# ---------------------------------------------------
# COMPUTE GINI INDEX
# ---------------------------------------------------
print("\n=== GINI INDEX RESULTS ===")

gini_weighted = gini_summary("Weighted Rating (all movies)", movies['weighted_rating'])
gini_vote = gini_summary("Vote Count (raw)", movies['vote_count'])
gini_logvote = gini_summary("Log1p Vote Count", np.log1p(movies['vote_count']))

# ---------------------------------------------------
# GINI FOR RECOMMENDED MOVIES
# ---------------------------------------------------
recommended_movie_ids = [19995, 140607, 135397, 76341, 87827]

rec_mask = movies['id'].isin(recommended_movie_ids)

if rec_mask.any():
    gini_recs = gini_summary("Weighted Rating (Recommended Movies)", movies.loc[rec_mask, 'weighted_rating'])
else:
    print("ERROR: Recommended movie IDs not found in dataset.")
```
